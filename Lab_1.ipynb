```python
#  1
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("C:/Users/38099/Desktop/Tasks/spotify-2023.csv", encoding='cp1251')
data = data.drop(574)
top_songs = data.sort_values(by='streams', ascending=False).head(10)

plt.figure(figsize=(10, 6))
plt.barh(top_songs['track_name'], top_songs['streams'], color='skyblue')
plt.xlabel('streams')
plt.ylabel('track_name')
plt.title('Top 10 Most Streamed Spotify Songs in 2023')
plt.gca().invert_yaxis()
plt.show()
```

```python
#  2
plt.figure(figsize=(8, 8))

plt.pie(top_songs['streams'], labels=top_songs['track_name'], autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Most Streamed Spotify Songs in 2023')
plt.axis('equal') 
plt.show()
```

```python
#  3
plt.figure(figsize=(8, 8))
danceability_counts = data['danceability_%'].value_counts().head(10)
plt.pie(danceability_counts, labels=danceability_counts.index, autopct='%1.1f%%')
plt.title('Distribution of Danceability among Top Spotify Songs in 2023')
plt.axis('equal')
plt.show()
```

```python
#  4
track_name = top_songs.iloc[0]['track_name']

track_data = data[data['track_name'] == track_name]

plt.figure(figsize=(10, 6))
plt.plot(track_data['released_month'], track_data['streams'], marker='o', linestyle='-', color='b')
plt.xlabel('')
plt.ylabel(' ')
plt.title(f'    "{track_name}"  2023 ')
plt.grid(True)
plt.xticks(range(1, 13))
plt.show()
```

```python
#  5
top_artists = data.groupby('artist(s)_name').sum()['streams'].sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
plt.bar(top_artists.index, top_artists.values, color='lightgreen')
plt.xlabel('artist(s)_name')
plt.ylabel('streams')
plt.title('Top 10 Most Popular Spotify Artists in 2023')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#  6
top_artists = data.groupby('artist(s)_name').sum()['streams'].sort_values(ascending=False).head(10)

#   
plt.figure(figsize=(10, 6))
plt.bar(top_artists.index, top_artists.values, color='skyblue')
plt.xlabel('')
plt.ylabel(' ')
plt.title('-10    Spotify  2023 ')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#  6
tracks = ['vampire', 'Cruel Summer', 'Sprinter']

selected_tracks = data[data['track_name'].isin(tracks)]

plt.figure(figsize=(10, 6))
plt.barh(selected_tracks['track_name'], selected_tracks['streams'], color=['skyblue', 'salmon', 'lightgreen'])
plt.xlabel(' ')
plt.ylabel('')
plt.title('      2023 ')
plt.tight_layout()
plt.show()
```
