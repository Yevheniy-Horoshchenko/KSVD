```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("shopping_trends.csv")

print("Розмір таблиці:", data.shape)

print("Пропуски в даних:\n", data.isnull().sum())

print("Кількість дублікатів:", data.duplicated().sum())

print("Типи даних стовпців:\n", data.dtypes)
```

```python
data.columns = data.columns.str.lower().str.replace(' ', '_')

# Побудова гістограми віку
plt.figure(figsize=(10, 6))
sns.histplot(data=data, x='age', bins=20, kde=True)
plt.title('Гістограма віку')
plt.xlabel('Вік')
plt.ylabel('Частота')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='item_purchased', order=data['item_purchased'].value_counts().index)
plt.title('Порівняння розподілу покупок за статтею')
plt.xlabel('Товар')
plt.ylabel('Кількість покупок')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='age', y='purchase_amount_(usd)')
plt.title('Залежність між віком та сумою покупок')
plt.xlabel('Вік')
plt.ylabel('Сума покупок (USD)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=data, x='review_rating', bins=10, kde=True)
plt.title('Розподіл оцінок товарів')
plt.xlabel('Оцінка товару')
plt.ylabel('Частота')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='gender')
plt.title('Розподіл покупок за статтю')
plt.xlabel('Стать')
plt.ylabel('Кількість покупок')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='review_rating', y='purchase_amount_(usd)')
plt.title('Залежність між рейтингом і сумою покупок')
plt.xlabel('Рейтинг товару')
plt.ylabel('Сума покупки (USD)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='season', order=data['season'].value_counts().index)
plt.title('Сезонність покупок')
plt.xlabel('Сезон')
plt.ylabel('Кількість покупок')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='size', order=data['size'].value_counts().index)
plt.title('Розподіл розмірів одягу')
plt.xlabel('Розмір')
plt.ylabel('Кількість покупок')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='payment_method')
plt.title('Кількість покупок відносно типу оплати')
plt.xlabel('Тип оплати')
plt.ylabel('Кількість покупок')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='previous_purchases', y='purchase_amount_(usd)')
plt.title('Залежність між кількістю попередніх покупок і сумою поточної покупки')
plt.xlabel('Кількість попередніх покупок')
plt.ylabel('Сума покупки (USD)')
plt.show()
```

```python
# завдання 2
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(x**2 + y**2)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
surface = ax.plot_surface(x, y, z, cmap='viridis')

ax.set_title('Залежність між трьома числовими змінними')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

fig.colorbar(surface, shrink=0.5, aspect=5)

plt.show()
```
